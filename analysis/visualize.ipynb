```python
import matplotlib.pyplot as plt
def addlabels(x,y, label):
    for i in range(len(x)):
        plt.text(i, y[i], label[i], ha = 'center')
plt.rcParams.update({'font.size': 14})
plt.figure(figsize=(10, 3))
clip = [0.4223, 0.4190, 0.3967, 0.4063, 0.4188, 0.4204, 0.4301]
models = ['stable diffusion', 'salun - van gogh', 'salun - nudity', 'salun - both', 'esd - van gogh', 'esd - nudity', 'esd - both']
color = ['tab:blue', 'tab:orange', 'tab:orange', 'tab:orange', 'tab:pink', 'tab:pink', 'tab:pink']
addlabels(models, clip, clip)
plt.bar(x=models, height=clip, color=color)
plt.xticks(rotation=10)
plt.ylim(0, 0.47)
plt.title("CLIP Score Evaluation of Prompt Alignment of Generated Images")
plt.ylabel("CLIP Score")
plt.xlabel('Model')
plt.show()
```

```python
import numpy as np
q16 = [456, 434, 138, 138, 416, 339, 344]
plt.figure(figsize=(10, 3))

q16 = [np.round(x/4703*10, 3) for x in q16]
models = ['stable diffusion', 'salun - van gogh', 'salun - nudity', 'salun - both', 'esd - van gogh', 'esd - nudity', 'esd - both']
color = ['tab:blue', 'tab:orange', 'tab:orange', 'tab:orange', 'tab:pink', 'tab:pink', 'tab:pink']
q16_str = [f'{x}%' for x in q16]
addlabels(models, q16, q16_str)
plt.bar(x=models, height=q16, color=color)
plt.xticks(rotation=20)
plt.ylim(0, 1.05)
plt.title("Q16 Classifier for Inappropriate Content in Generated Images")
plt.ylabel("Percent Inappropriate")
plt.xlabel('Model')
plt.show()
```
